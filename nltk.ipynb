```python
from nltk.tokenize import word_tokenize # Tokenization
from nltk.corpus import stopwords # Stopword removal
from nltk.stem import PorterStemmer # Stemming
from nltk.stem import WordNetLemmatizer # Lemmatization
```

```python
raw_text = "I am learning python programming, Python is greatest language..!!"
```

```python
# step 1: lowercase
raw_text = raw_text.lower()
print("After converting to lowercase:", raw_text)
```

```python
#step 2: tokenization
print(raw_text.split())
```

```python
tokens = word_tokenize(raw_text)
print("After tokenization:", tokens)
```

```python
#step 3: stopword removal
english_stopwords = stopwords.words('english')
```

```python
filtered_tokens = []
for word in tokens:
    if word not in english_stopwords:
        filtered_tokens.append(word)

print("After stopword removal:", filtered_tokens)
```

```python
# Step 4: removing punctuation
import string
punctuations = string.punctuation
print("Punctuations available:", punctuations)
```

```python
clean_tokens = [word for word in filtered_tokens if word not in punctuations]
print("After removing punctuation:", clean_tokens)
```

```python
stem = PorterStemmer()
#stem.stem("playing")
#stem.stem("played")
#stem.stem("watching")
#stem.stem("flying")
stem.stem("bought")
```

```python
wnet = WordNetLemmatizer()
#wnet.lemmatize("running", pos='v')
#wnet.lemmatize("bought", pos='v')
wnet.lemmatize("went", pos='v')
```

```python
lemmatize_words = []
for word in clean_tokens:
    lemmatize_words.append(wnet.lemmatize(word, pos='v'))
print("After lemmatization:", lemmatize_words)
```

```python
final_tokens = []
for word in lemmatize_words:
    if word.isalpha():
        final_tokens.append(word)
print("Final cleaned tokens:", final_tokens)       
```
